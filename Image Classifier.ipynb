# Deep Learning

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout
```

```python
model = Sequential()
```

```python
# Conv2D fliter no. =16 //fliter size= 3X3 pixels //movement= 1 stride at a time and repeat process until it scans the whole image
# relu activation= converts any negative number of the image from 0 upwards, and positive int remains unchanged
# shape input= images are in 256px high, 256px wide by 3 channels(RBG/BGR)  
# MaxPooling2D takes the max value after the model activation and condenses the info
model.add(Conv2D(16, (3,3), 1, activation='relu', input_shape=(256,256,3)))
model.add(MaxPooling2D())

# Conv2D fliter no. =32 //fliter size= 3X3 pixels //movement= 1 stride at a time and repeat process until it scans the whole image
# relu activation= converts any negative number of the image from 0 upwards, and positive int remains unchanged
model.add(Conv2D(32, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())

model.add(Conv2D(16, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())

model.add(Flatten())

# relu activation with a dense layer of 256 neurons
# sigmoid activation converts numbers in a range of 0-1 and a single dense layer
model.add(Dense(256, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
#use the Adam optimizer, specify our loss using binary classification model, using accuracy as a metric to track the models accuracy 
#which is how accrate the model is classifying as eith 0 or 1
model.compile('adam', loss=tf.losses.BinaryCrossentropy(), metrics=['accuracy'])
```

```python
model.summary()
```

Train The Model

```python
#logs director to see performance of the model
logdir='logs'
```

```python
#callback for saving models at a particular checkpoint
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=logdir)
```

```python
#fit Used for training a machine learning model.
hist = model.fit(train, epochs=20, validation_data=val, callbacks=[tensorboard_callback])
```

```python
fig = plt.figure()
plt.plot(hist.history['loss'], color='blue', label='loss')
plt.plot(hist.history['val_loss'], color='red', label='val_loss')
fig.suptitle('Loss', fontsize=20)
plt.legend(loc="upper left")
plt.show()
```

```python
fig = plt.figure()
plt.plot(hist.history['accuracy'], color='blue', label='accuracy')
plt.plot(hist.history['val_accuracy'], color='orange', label='val_accuracy')
fig.suptitle('Accuracy', fontsize=20)
plt.legend(loc="upper left")
plt.show()
```

# Evaluation of Performance

```python
from tensorflow.keras.metrics import Precision, Recall, BinaryAccuracy
```

```python
pre = Precision()
re = Recall()
acc = BinaryAccuracy()
```

```python
len(test)
```

```python
#looping through each batch in the testing data
for batch in test.as_numpy_iterator(): 
    X, y = batch
    yhat = model.predict(X)
    pre.update_state(y, yhat)
    re.update_state(y, yhat)
    acc.update_state(y, yhat)
```

```python
print(f'Precisio:{pre.result().numpy()}, Recall:{re.result().numpy()}, Accuracy:{acc.result().numpy()}')
```

```python
def predict_class(img):
    C_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    resize = tf.image.resize(C_img, (256,256))
    plt.imshow(resize.numpy().astype(int))
    yhat = model.predict(np.expand_dims(resize/255, 0))
    plt.show()
    if yhat > 0.5: 
        print(f'Predicted class is a Cat')
    else:
        print(f'Predicted class is Dog')
```
